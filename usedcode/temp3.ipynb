```python
from pymatgen.core import Structure, Species
from mp_api.client import mprester
```

```python
li2mno3 = Structure.from_file("data/POSCAR_li2mno3")
```

```python
li2mno3.remove_species(["O"])
```

```python
neighbors = li2mno3.get_all_neighbors(r= 2.93998)
```

```python
r_neighbors = 2.0
while r_neighbors <4.0:
    neighbors = li2mno3.get_all_neighbors(r=r_neighbors)
    if len(neighbors[0]) < 10:
        r_neighbors+=0.01
    elif len(neighbors[0]) ==10:
        print("meeting the requirement")
        print("r={0}".format(r_neighbors))
        break
    elif len(neighbors[0]) > 10:
        print("Seem a smaller step is needed")
        break
```
