# éç›‘ç£å­¦ä¹ æ¦‚è¿°

$\quad$ä¸ç›‘ç£å­¦ä¹ ä¸åŒ, **éç›‘ç£å­¦ä¹ **(unsupervised learning)æ—¨åœ¨æŒ–æ˜é«˜ç»´æ•°æ®åˆ†å¸ƒä¸­çš„å†…åœ¨ç»“æ„, æ‰€ä½¿ç”¨çš„è®­ç»ƒæ•°æ®é›†$\mathcal{D} = \{\mathbf{x}^{(n)}\}_{n=1}^N$åªæœ‰ç‰¹å¾, æ²¡æœ‰â€œæ ‡ç­¾â€(æˆ–è¯´â€œæ­£ç¡®ç­”æ¡ˆâ€). ä»æ¦‚ç‡è§†è§’æ¥çœ‹, ç›‘ç£å­¦ä¹ ä¸éç›‘ç£å­¦ä¹ çš„åŒºåˆ«åœ¨äº:
- ç›‘ç£å­¦ä¹ æ‰€æ‹Ÿåˆçš„æ¦‚ç‡åˆ†å¸ƒä¸º**æ¡ä»¶åˆ†å¸ƒ**$p(\mathbf{t}|\mathbf{x})$;
- éç›‘ç£å­¦ä¹ ç›´æ¥åˆ»ç”»æ•°æ®ç‰¹å¾æœ¬èº«çš„åˆ†å¸ƒ$p(\mathbf{x})$.

$\quad$æˆ‘ä»¬å°†è®¨è®ºä¸¤ç§å…¸å‹çš„éç›‘ç£å­¦ä¹ ä»»åŠ¡, åˆ†åˆ«æ˜¯:
- **é™ç»´**(dimensionality reduction), å­¦ä¹ ä»é«˜ç»´æ•°æ®$\mathbf{x} \in \mathbb{R}^D$åˆ°ä½ç»´æ•°æ®$\mathbf{z} \in \mathbb{R}^M$çš„(å‚æ•°åŒ–)æ˜ å°„$f(\mathbf{x}; \boldsymbol{\theta})$, ä½¿å¾—å°½å¯èƒ½å°‘åœ°æŸå¤±é«˜ç»´ç©ºé—´ä¸­çš„ä¿¡æ¯;
  - å¸¸è§åº”ç”¨åœºæ™¯: æ•°æ®å¯è§†åŒ–ã€æé«˜ç®—æ³•æ•ˆç‡ä¸ç¨³å¥æ€§ã€â€¦â€¦
- **èšç±»**(clustering), åŸºäºæ•°æ®æ ·æœ¬$\mathbf{x} \in \mathbb{R}^D$ä¹‹é—´çš„ç›¸ä¼¼æ€§, å¯¹å„æ ·æœ¬åˆ†é…ç±»åˆ«æ ‡ç­¾$z$, ç›¸ä¼¼æ ·æœ¬å½’ä¸ºä¸€ç±».
  - å¸¸è§åº”ç”¨åœºæ™¯: ç¾¤ä½“çš„åˆ†å‰²ã€æ–°é—»èšåˆã€è›‹ç™½æ„è±¡çš„æå–ã€â€¦â€¦

ä¹Ÿå¯ä»¥åœ¨ç»Ÿä¸€çš„æ¡†æ¶ä¸‹è®¤è¯†è¿™ä¸¤ç§ä»»åŠ¡: éç›‘ç£å­¦ä¹ æ¨¡å‹è¯•å›¾å­¦ä¹ æŸç§éšå˜é‡$\mathbf{z}$. èšç±»ç®—æ³•ä¸­çš„éšå˜é‡æ˜¯ç¦»æ•£çš„ç±»åˆ«æ ‡ç­¾, é™ç»´ç®—æ³•ä¸­çš„éšå˜é‡åˆ™æ˜¯æ•°æ®åœ¨ä¸€ä¸ªä½ç»´è¿ç»­ç©ºé—´ä¸­çš„**è¡¨ç¤º**(representation).

# é™ç»´ç®—æ³•

$\quad$æœ¬èŠ‚æ€»å‡å®šè¾“å…¥æ•°æ®$\mathbf{X}$åšäº†ä¸­å¿ƒåŒ–å¤„ç†: å‡å»ç‰¹å¾çš„å‡å€¼, ä½¿å¾—æ¯ä¸ªç‰¹å¾çš„æ–°å‡å€¼ä¸º0.

## ä¸»å…ƒåˆ†ææ³•

$\quad$æœ€ç®€å•çš„æ˜ å°„å‡½æ•°æ˜¯çº¿æ€§å‡½æ•°! å€Ÿç”¨**è‡ªç¼–ç å™¨**(autoencoder)æ¨¡å‹ä¸­çš„æœ¯è¯­, æˆ‘ä»¬å–**ç¼–ç å™¨**(encoder function, ç”¨äºæ•°æ®é™ç»´)$f_\mathrm{E}(\mathbf{x}) := \mathbf{W}^\mathrm{T}\mathbf{x}$, **è§£ç å™¨**(decoder function, ç”¨äºä»ä½ç»´æ•°æ®é‡æ„å‡ºé«˜ç»´æ•°æ®)$f_\mathrm{D}(\mathbf{z}) := \mathbf{W}\mathbf{z}$, å¹¶è¯•å›¾æœ€å°åŒ–é‡æ„è¯¯å·®. äºæ˜¯, å®šä¹‰â€œæŸå¤±å‡½æ•°â€ä¸ºä¸‹è¿°çš„**ç•¸å˜å‡½æ•°**(distortion function):
$$\begin{aligned}
\mathcal{L}(\mathbf{W}) &:= \frac{1}{N}\sum_{n=1}^N{\|\mathbf{x}^{(n)} - f_\mathrm{D}(f_\mathrm{E}(\mathbf{x}^{(n)}))\|^2}\\
&= \frac{1}{N}\|\mathbf{X} - \mathbf{Z}\mathbf{W}^\mathrm{T}\|_F^2,
\end{aligned}$$
å…¶ä¸­, $\|\cdot\|_F$ä¸ºçŸ©é˜µçš„**FrobeniusèŒƒæ•°**(Frobenius norm), å®šä¹‰ä¸ºæ‰€æœ‰çŸ©é˜µå…ƒçš„å¹³æ–¹å’Œçš„ç®—æœ¯å¹³æ–¹æ ¹. è¿™å°±æ˜¯**ä¸»å…ƒåˆ†æ**(principal component analysis, PCA)çš„åŸºæœ¬æ€æƒ³.

- ç†è®ºè¯¾ä¸Šå·²ç»è¯æ˜, æœ€ä¼˜çŸ©é˜µ$\mathbf{W}^* \in \mathbb{R}^{D \times M}$åº”å–ä¸ºåæ–¹å·®çŸ©é˜µ$\Sigma := \mathbf{X}^\mathrm{T}\mathbf{X}$çš„å‰$M$ä¸ªä¸»æœ¬å¾å‘é‡$\mathbf{U}_M$. å³: **PCAä¾é æ•°æ®ç©ºé—´çš„æ—‹è½¬å˜æ¢ä¸æŠ•å½±é™ç»´, æœ€å¤§ç¨‹åº¦åœ°ä¿ç•™æ•°æ®æ–¹å·®**.
  - ç®—æ³•å®è·µä¸­, æ›´ç¨³å¥çš„åšæ³•æ˜¯æ„é€ **ç›¸å…³çŸ©é˜µ**(correlation matrix)è€Œéåæ–¹å·®çŸ©é˜µ, è¿™ç›¸å½“äºåšäº†æ•°æ®çš„é‡æ ‡åº¦.
- ä»æ¦‚ç‡è®ºè§†è§’, PCAé—®é¢˜è¿˜å¯ä»¥ç­‰ä»·åœ°åšå¦‚ä¸‹æè¿°: æˆ‘ä»¬å‡è®¾éšå˜é‡$\mathbf{z}$æœä»æ ‡å‡†æ­£æ€åˆ†å¸ƒ, è€Œä»$\mathbf{z}$å‡ºå‘é‡æ„å‡ºçš„$\mathbf{x}$æœä»ä»¥$\hat{\mathbf{x}} = \mathbf{W}\mathbf{z}$ä¸ºä¸­å¿ƒçš„æ¡ä»¶Gaussianåˆ†å¸ƒ. å³:
  $$\begin{aligned}
  p(\mathbf{z}) &= \mathcal{N}(\mathbf{z} | \mathbf{0}, \mathbf{I}),\\
  p(\mathbf{x} | \mathbf{z}, \boldsymbol{\theta}) &= \mathcal{N}(\mathbf{x} | \mathbf{W}\mathbf{z}, \sigma^2\mathbf{I}).\\
  \Rightarrow p(\mathbf{x} | \boldsymbol{\theta}) &= \int{\mathrm{d}\mathbf{z}p(\mathbf{z})p(\mathbf{x} | \mathbf{z}, \boldsymbol{\theta})}\\
  &= \mathcal{N}(\mathbf{x} | \mathbf{0}, \mathbf{W}\mathbf{W}^\mathrm{T} + \sigma^2\mathbf{I}).
  \end{aligned}$$
  æ­¤æ—¶, å¯ä»¥è¯æ˜, æå¤§ä¼¼ç„¶ä¼°è®¡å°†ç»™å‡ºä¸‹è¿°å½¢å¼çš„æœ€ä¼˜è§£:
  $$
  \mathbf{W}^* = \arg{\min_\mathbf{W}{\mathrm{NLL}(\mathbf{W}; \sigma^2)}} = \mathbf{U}_M(\mathbf{\Lambda}_M - \sigma^2\mathbf{I})^{\frac{1}{2}},
  $$
  å…¶ä¸­, $\mathbf{U}_M$ä¸ºæ•°æ®$\mathbf{x}$çš„åæ–¹å·®çŸ©é˜µ$\Sigma$çš„å‰$M$ä¸ªä¸»æœ¬å¾å‘é‡, $\mathbf{\Lambda}$ä¸ºå‰$M$ä¸ªä¸»æœ¬å¾å€¼æ„æˆçš„å¯¹è§’é˜µ, ä¹Ÿå³:
  $$
  \mathbf{\Sigma} = \mathbf{U}^\mathrm{T}\mathbf{\Lambda}\mathbf{U}.
  $$
  çš„å·¦ä¸Šæ–¹$M \times M$å­å—.
  - PCAæ“ä½œä½¿å¾—åæ–¹å·®çŸ©é˜µå¯¹è§’åŒ–, æ¶ˆé™¤å„ä¸ªåˆ†é‡ä¹‹é—´çš„ç›¸å…³æ€§.
  - è‹¥å–é›¶å™ªå£°æé™$\sigma^2 \to 0$, åˆ™$\mathbf{W}^* = \mathbf{U}_M\mathbf{\Lambda}_M^{\frac{1}{2}}$æ°å¥½å°±æ˜¯PCAé—®é¢˜çš„è§£!

$\quad$è¯¸å¦‚PCAä¸€ç±»çš„é™ç»´ç®—æ³•å¾€å¾€åŸºäºçŸ©é˜µåˆ†è§£(ä¾‹å¦‚ä¸Šè¿°çš„æœ¬å¾å€¼åˆ†è§£), å®ƒä»¬éƒ½å°è£…åœ¨[`sklarn.decomposition`](https://scikit-learn.org/stable/modules/classes.html#module-sklearn.decomposition)æ¨¡å—ä¸­. ä»¥[`PCA`](https://scikit-learn.org/stable/modules/generated/sklearn.decomposition.PCA.html#sklearn-decomposition-pca)ä¸ºä¾‹, ä¸»è¦çš„è°ƒç”¨æ¥å£åŒ…æ‹¬:
- æ¨¡å‹åˆ›å»º`pca = PCA()`, éœ€ä¼ å…¥å‚æ•°`n_components`, ä»£è¡¨éšç©ºé—´ç»´æ•°$M$.
  - é»˜è®¤æƒ…å½¢ä¸‹å–$M=D$, å³: åªå¯¹ç‰¹å¾ä½œâ€œæ—‹è½¬å˜æ¢â€ä½†ä¸ä½œä½ç»´æˆªæ–­ã€ä¿ç•™æ‰€æœ‰ç»´åº¦.
- æ¨¡å‹è®­ç»ƒ`pca.fit()`, åŸºäº[**å¥‡å¼‚å€¼åˆ†è§£**(singular value decomposition, SVD)](https://en.wikipedia.org/wiki/Singular_value_decomposition)è¿›è¡Œé™ç»´å˜æ¢$\mathbf{W}^*$çš„æ±‚è§£.
  - æ³¨æ„: æ¨èå¯¹æ•°æ®é¢„å…ˆæ‰‹åŠ¨ä½œå‡å€¼-æ–¹å·®å½’ä¸€åŒ–. è¿™ä¸€è®­ç»ƒæ–¹æ³•åªä¼šè‡ªåŠ¨å¯¹æ•°æ®ä½œä¸­å¿ƒåŒ–, ä½†ä¸ä¼šé™¤ä»¥æ–¹å·®(ä¸ä½œé‡æ ‡åº¦).
- é™ç»´å˜æ¢çš„è°ƒç”¨`pca.transform()`, ç”¨è®­ç»ƒå¥½çš„$\mathbf{W}^*$è¿›è¡Œé™ç»´å˜æ¢$\mathbf{z} = (\mathbf{W}^*)^\mathrm{T}\mathbf{x}$.
  - ä¹Ÿå¯ä»¥(å¯¹è®­ç»ƒé›†)ä½¿ç”¨å…ˆè®­ç»ƒåå˜æ¢çš„æ¥å£`fit_transform()`.

```python
# Example: iris dataset
from matplotlib import pyplot as plt
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA

iris = load_iris()
pca = PCA(n_components=3)
X_iris_reduced = pca.fit_transform(iris.data)

fig = plt.figure(1, figsize=(8, 6))
ax = fig.add_subplot(111, projection="3d", elev=-150, azim=110)
ax.scatter(
    X_iris_reduced[:, 0],
    X_iris_reduced[:, 1],
    X_iris_reduced[:, 2],
    c=iris.target,
    s=40,
)
ax.set_title("First three PCA dimensions")
ax.set_xlabel("PC1")
ax.xaxis.set_ticklabels([])
ax.set_ylabel("PC2")
ax.yaxis.set_ticklabels([])
ax.set_zlabel("PC3")
ax.zaxis.set_ticklabels([])
plt.show()
```

- å„ä¸ªä¸»å…ƒæ‰€èƒ½è§£é‡Šçš„æ–¹å·®(å¯¹åº”çš„å„ä¸ªæœ¬å¾å€¼)å¯ç”±æ¨¡å‹çš„ç±»å±æ€§`explained_variance`æˆ–`explained_variance_ratio_`è®¿é—®;
  - ç”±äºåæ–¹å·®çŸ©é˜µæ˜¯**æ­£å®šçŸ©é˜µ**(positive definite matrix), åˆ™...
    - è‹¥$M=D$, åˆ™$\mathbf{W}^*$ä¸ºæ­£äº¤çŸ©é˜µ(ä¿è¿¹), å„ä¸ªæœ¬å¾å€¼çš„å’Œ(ä¸»å…ƒæ–¹å·®å’Œ)åº”ç­‰äºåŸå§‹æ•°æ®çš„æ€»æ–¹å·®ï¼›
    - æ¯ä¸ªæœ¬å¾å€¼å°†æ•æ‰æ€»æ–¹å·®çš„ä¸€éƒ¨åˆ†, è°“ä¹‹â€œè§£é‡Šâ€.
  - å„ä¸ªæœ¬å¾å€¼æŒ‰é™åºæ’åˆ—è¾“å‡º.
- å˜æ¢çŸ©é˜µ$\mathbf{W}^*$å¯ç”±æ¨¡å‹çš„ç±»å±æ€§`components_`è®¿é—®.
  - ç³»æ•°åæ˜ äº†åŸç‰¹å¾åˆ†é‡ä¸æ–°ä¸»å…ƒåˆ†é‡ä¹‹é—´çš„â€œæŠ•å½±â€å…³ç³», ä»è€Œæè¿°äº†å„ä¸ªä¸»å…ƒä¸ŠåŸç‰¹å¾çš„â€œç»„æˆæˆåˆ†â€/å„ä¸ªç‰¹å¾ä¸Šä¸»å…ƒçš„â€œç»„æˆæˆåˆ†â€.
  - å½¢çŠ¶ä¸º`(n_components, n_features)`, äº‹å®ä¸Šä¸º$(\mathbf{W}^*)^\mathrm{T}$. å„è¡ŒæŒ‰æœ¬å¾å€¼é™åºæ’åˆ—è¾“å‡º.

## éçº¿æ€§é™ç»´ä¸æµå½¢å­¦ä¹ 

$\quad$ä¸ºä»€ä¹ˆæƒ³è¦é™ç»´? æ—¥å¸¸è§‚å¯Ÿä¼¼ä¹è¡¨æ˜, å¤šæ•°â€œè‡ªç„¶äº§ç”Ÿçš„â€é«˜ç»´æ•°æ®æ ·æœ¬æ€»å­˜åœ¨ç‰¹å®šçš„ç»“æ„å’Œè§„å¾‹, è¿™äº›ç»“æ„åªéœ€è¦åœ¨è¾ƒä½ç»´åº¦ä¸‹å°±è¶³ä»¥æè¿°. ä¾‹å¦‚, æ‰‹å†™æ•°å­—6çš„å›¾ç‰‡åŒ…å«64x64 = 4096ç»´åº¦çš„åƒç´ , ä½†è¿™äº›åƒç´ ä¹‹é—´é«˜åº¦ç›¸å…³, å†…ç¦€çš„ä½ç»´ç»“æ„å¯ä»¥(ç”¨æ–‡å­—)æè¿°ä¸ºâ€œæ•°å­—6â€. æè¿°è¿™ç§â€œç»“æ„â€çš„æ•°å­¦è¯­è¨€æ˜¯**æµå½¢**(manifold).
- ç²—æµ…åœ°ç†è§£, â€œæµå½¢â€å°±æ˜¯æ‹¥æœ‰å±€åŸŸâ€œå¹³é¢â€ç»“æ„çš„ä¸€ä¸ªä½ç»´ç©ºé—´.
  - ä¾‹å¦‚: åœ°çƒè¡¨é¢å°±å¯ä»¥ç²—ç•¥åœ°çœ‹ä½œä¸€ä¸ªæµå½¢. è¿™æ˜¯ä¸€ä¸ªå­˜åœ¨äºä¸‰ç»´ç©ºé—´ä¸­çš„äºŒç»´æ›²é¢, è€Œåœ¨è¡¨é¢ä¸Šä»»ä¸€ç‚¹çš„é™„è¿‘ä¸€å—å°åŒºåŸŸå†…, åœ°è¡¨çœ‹èµ·æ¥éƒ½æ˜¯å¹³å¦çš„.
  - ç§°ä¸€ä¸ª$d$-ç»´ç©ºé—´$\mathcal{X}$ä¸º**æµå½¢**, è‹¥å¯¹ä»»æ„$x \in \mathcal{X}$éƒ½å­˜åœ¨ä¸€ä¸ªä¸$\mathbb{R}^d$ç©ºé—´(æ‹“æ‰‘ä¸Š)ç­‰ä»·çš„é‚»åŸŸ$U(x)$. è¿™ä¸ªç­‰ä»·çš„ç©ºé—´$\mathbb{R}^d$ç§°ä¸º**æ­£åˆ‡ç©ºé—´**(tangent space), è®°ä½œ$\mathcal{T}_x := T_x\mathcal{X}$.
- **æµå½¢å‡è®¾**(manifold hypothesis): å¯¹ç»ç”±è‡ªç„¶è§„å¾‹äº§ç”Ÿçš„(é«˜ç»´)æ•°æ®, æˆ‘ä»¬æ€»å‡è®¾å®ƒä»¬åˆ†å¸ƒäºä¸€ä¸ªä½ç»´æµå½¢ä¸Š.

$\quad$å¯¹æ•°æ®è¿›è¡Œéçº¿æ€§é™ç»´çš„ä»»åŠ¡å¾€å¾€ä»¥å­¦ä¹ ä½ç»´æµå½¢ç»“æ„ä¸ºç›®æ ‡, å› æ­¤ç§°ä¸º**æµå½¢å­¦ä¹ **(manifold learning). ä¸æµå½¢å­¦ä¹ ç›¸å…³çš„ç®—æ³•å°è£…äº[`sklearn.manifold`](https://scikit-learn.org/stable/modules/classes.html#module-sklearn.manifold)æ¨¡å—ä¸­. æˆ‘ä»¬æœ¬èŠ‚ä»¥ä¸‰ç»´ç©ºé—´ä¸­çš„[Swiss-rollæ•°æ®é›†](https://scikit-learn.org/stable/modules/generated/sklearn.datasets.make_swiss_roll.html#sklearn-datasets-make-swiss-roll)ä¸ºä¾‹æ¼”ç¤ºå‡ ç§æµå½¢ç®—æ³•çš„å­¦ä¹ æ•ˆæœ. è¿™æ˜¯ä¸€ä¸ªåœ¨èšç±»ã€é™ç»´ã€ç”Ÿæˆæ¨¡å‹ä¸­ç›¸å½“å¸¸ç”¨çš„ä½œä¸ºâ€œç©å…·â€çš„äººå·¥åˆæˆæ•°æ®é›†, æ—¨åœ¨å®šæ€§æ¼”ç¤ºä¸åˆæ­¥æ£€æŸ¥.

```python
from matplotlib.ticker import NullFormatter
from sklearn.datasets import make_swiss_roll

def plot_swiss_roll_dataset(X, color, noise):
    fig = plt.figure(1, figsize=(8, 6))
    ax = fig.add_subplot(111, projection="3d")
    ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=color, cmap=plt.cm.Spectral)
    ax.view_init(4, -72)
    ax.xaxis.set_major_formatter(NullFormatter())
    ax.yaxis.set_major_formatter(NullFormatter())
    ax.zaxis.set_major_formatter(NullFormatter())
    ax.set_title(f"Swiss roll dataset with noise {noise}")
    ax.axis("tight")
    plt.show()
```

```python
n_samples = 1000
X, color = make_swiss_roll(n_samples=n_samples, noise=0, random_state=42)
plot_swiss_roll_dataset(X, color, noise=0)
```

- ä¸€ä¸ªæœ€æ—©çš„æµå½¢å­¦ä¹ ç®—æ³•ä¸º**ç­‰åº¦é‡æ˜ å°„**(isometric mapping). æˆ‘ä»¬é‡æ–°å®šä¹‰â€œè·ç¦»â€çš„åº¦é‡æ–¹å¼: **æµ‹åœ°çº¿è·ç¦»**(geodesic distance), å¹¶åœ¨è¯¥è·ç¦»åº¦é‡ä¸‹æ²¿ç”¨ç±»ä¼¼PCAçš„é™ç»´ç®—æ³•, ä½¿å¾—: *å­¦ä¹ åˆ°çš„ä½ç»´è¡¨ç¤ºå°½å¯èƒ½ä¿æŒæµ‹åœ°çº¿è·ç¦»ä¸å˜*.
  - æµ‹åœ°çº¿è·ç¦»å®šä¹‰ä¸º(å¸¦æƒ)è¿‘é‚»å›¾ä¸Šçš„æœ€çŸ­è·¯å¾„é•¿. å®ƒä¸æ™®é€šEuclideanè·ç¦»çš„æœ€é‡è¦çš„åŒºåˆ«æ˜¯*è€ƒè™‘äº†æ›²é¢ç»“æ„å¼•èµ·çš„äº’é€šè·¯å¾„å˜åŒ–*.
  - ä¸¤ä¸ªEuclideanè·ç¦»ç›¸è¿‘çš„ç‚¹ä»ç„¶å¯èƒ½å…·æœ‰å¾ˆè¿œçš„æµ‹åœ°çº¿è·ç¦», å› ä¸ºå®ƒä»¬åœ¨ä½ç»´æµå½¢ä¸Šéœ€è¦è·¨è¶Šå¾ˆé•¿çš„è·ç¦»æ‰èƒ½äº’ç›¸åˆ°è¾¾.
- ç­‰åº¦é‡æ˜ å°„å¯ç”±[`sklearn.manifold.Isomap`](https://scikit-learn.org/stable/modules/generated/sklearn.manifold.Isomap.html#sklearn-manifold-isomap)æ¨¡å—å®ç°. è°ƒç”¨æ¥å£ä»ç„¶æ˜¯`fit()`ã€`transform()`ä¸ç»„åˆç‰ˆæœ¬`fit_transform()`.
- ç­‰åº¦é‡æ˜ å°„æ˜¯**æ‹“æ‰‘ä¸ç¨³å®šçš„**(topological instability), åŸå§‹æ•°æ®ä¸Šçš„å™ªå£°å¾®æ‰°å¾ˆå®¹æ˜“å¼•èµ·è¿‘é‚»å›¾çš„â€œå˜å½¢â€, ä»è€Œç»™å‡ºä¸åˆç†çš„ç»“æœ.
  - ä½ å¯ä»¥è‡ªè¡Œå°è¯•ä¿®æ”¹`n_neighbors`å‚æ•°ä»¥å‡å°æ„å»ºè¿‘é‚»å›¾æ—¶çš„é‚»å±…æ•°, ä½†è¿™ä¸èƒ½å¾ˆå¥½åœ°è§£å†³æ‹“æ‰‘ä¸ç¨³å®šæ€§ğŸ˜¥, å› ä¸ºå­¦ä¹ åˆ°çš„æµå½¢å°†å‘ˆç°å‡ºâ€œå½¼æ­¤å­¤ç«‹â€çš„è‹¥å¹²å°åŒºåŸŸ, éš¾ä»¥æ³›åŒ–.

```python
def plot_learned_manifold(X_reduced, color, method, noise):
    fig = plt.figure()
    ax = fig.add_subplot(111)
    ax.scatter(X_reduced[:, 0], X_reduced[:, 1], c=color, cmap=plt.cm.Spectral)
    ax.set_title(method + f" with noise {noise}")
    ax.xaxis.set_major_formatter(NullFormatter())
    ax.yaxis.set_major_formatter(NullFormatter())
    ax.axis("tight")
    plt.show()
```

- SNE(stochastic neighbor embedding)ç®—æ³•å°†æ ·æœ¬ç›¸ä¼¼æ€§å»ºæ¨¡ä¸ºç”±Euclideanè·ç¦»å†³å®šçš„æ¡ä»¶æ¦‚ç‡. å¸¸ç§°ä¸º**äº²å’Œæ€§**(affinity).
  - å¯¹ç»™å®šæ ·æœ¬$i$, å…¶é€‰æ‹©æ ·æœ¬$j$ä½œä¸ºè¿‘é‚»ç‚¹çš„æ¦‚ç‡å¸¸å®šä¹‰ä¸ºGaussianåˆ†å¸ƒ:
    $$
    p_{j | i} = \frac{
    \exp{\left\{-\frac{1}{2\sigma_i^2}\|\mathbf{x}^{(i)} - \mathbf{x}^{(j)}\|^2\right\}}
    }{
    \sum_{k \ne i}{\exp{\left\{-\frac{1}{2\sigma_i^2}\|\mathbf{x}^{(i)} - \mathbf{x}^{(k)}\|^2\right\}}}
    },
    $$
    å…¶ä¸­, $\sigma_i^2$ä»£è¡¨æ ·æœ¬$i$çš„â€œå¸¦å®½â€(band width), ä»£è¡¨è¿‘é‚»å…³ç³»çš„å®¹å¿åº¦($\sigma_i^2$è¶Šå¤§, å®¹å¿åº¦è¶Šé«˜, è¶Šæœ‰å¯èƒ½å…è®¸è¾ƒè¿œçš„æ ·æœ¬ç‚¹è¢«é€‰ä¸ºè¿‘é‚»ç‚¹), ä¸â€œè¿‘é‚»æ•°â€èµ·ç±»ä¼¼çš„ä½œç”¨.
  - éšç©ºé—´ä¸Š, æˆ‘ä»¬ä½œç±»ä¼¼çš„å®šä¹‰:
    $$
    q_{j | i} = \frac{
    \exp{\left\{-\|\mathbf{z}^{(i)} - \mathbf{z}^{(j)}\|^2\right\}}
    }{
    \sum_{k \ne i}{\exp{\left\{-\|\mathbf{z}^{(i)} - \mathbf{z}^{(k)}\|^2\right\}}}
    },
    $$
    åªä¸è¿‡ä¸å†å¼•å…¥å¸¦å®½å‚æ•°, å› ä¸ºä¸åŒçš„å¸¦å®½å‚æ•°åªæ˜¯å¯¹å­¦ä¹ åˆ°çš„æµå½¢ç»“æ„ä½œâ€œç¼©æ”¾â€, ä¸å½±å“æ‹“æ‰‘å…³ç³».
  - ä¸€èˆ¬åœ°, SNEç®—æ³•è¯•å›¾ä¿æŒé«˜ç»´æ•°æ®ç©ºé—´ä¸ä½ç»´éšç©ºé—´ä¸Šçš„äº²å’Œæ€§, å³è°ƒæ•´éšç©ºé—´è¡¨ç¤º$\{\mathbf{z}^{(i)}\}_{i=1}^N$, ä»¥ä¼˜åŒ–å„ä¸ªæ ·æœ¬çš„æ€»KLæ•£åº¦:
    $$
    \mathcal{L} := \sum_{i=1}^N{\sum_{j=1}^N{p_{j | i}\log{\frac{p_{j | i}}{q_{j | i}}}}}.
    $$
- é€šå¸¸çš„SNEç®—æ³•é¢ä¸´â€œæ‹¥æŒ¤é—®é¢˜â€(crowding problem). æ³¨æ„éšç©ºé—´çš„æ¡ä»¶æ¦‚ç‡$q_{j | i}$è¡°å‡å¾—éå¸¸å¿«, æ‰€ä»¥æ ·æœ¬åœ¨ä½ç»´æµå½¢ä¸Šçš„åˆ†å¸ƒå°†éå¸¸â€œæ‹¥æŒ¤â€.
  - **t-SNEç®—æ³•**å°†éšç©ºé—´ä¸Šçš„äº²å’Œæ€§å»ºæ¨¡ä¸ºâ€œå°¾å·´æ›´é•¿â€ã€è¡°å‡æ›´æ…¢çš„[å­¦ç”Ÿtåˆ†å¸ƒ](https://en.wikipedia.org/wiki/Student%27s_t-distribution):
    $$
    q_{j | i} = \frac{
    (1 + \|\mathbf{z}^{(i)} - \mathbf{z}^{(j)}\|^2)^{-1}
    }{
    \sum_{k \ne i}{(1 + \|\mathbf{z}^{(i)} - \mathbf{z}^{(k)}\|^2)^{-1}}
    }.
    $$
  - ä»¥[`sklearn.manifold.TSNE`](https://scikit-learn.org/stable/modules/generated/sklearn.manifold.TSNE.html#sklearn.manifold.TSNE)æ¨¡å—è°ƒç”¨, â€œå¸¦å®½â€å‚æ•°æ ¹æ®`perplexity`æ§åˆ¶.
    - ä¸åŒçš„`perplexity`å€¼å¾€å¾€ä¼šæ˜¾è‘—å½±å“é™ç»´ç»“æœ, æ¨èè¿›è¡Œå¤šæ¬¡è¯•éªŒ.
- t-SNEç®—æ³•æ›´å–„äºæŒ–æ˜æ•°æ®çš„å±€åŸŸç»“æ„, å…¶ç»™å‡ºçš„æµå½¢ç»“æ„å¾€å¾€å‘ˆç°ä¸º**è‹¥å¹²ç»„å½¼æ­¤åˆ†éš”çš„ç°‡, æ¯ä¸ªç°‡å†…æ ·æœ¬åˆ†å¸ƒå¾—ååˆ†ç´§å¯†**.
  - Isomapæ›´é€‚åˆå­¦ä¹ å•ä¸€ã€è¿ç»­çš„æµå½¢ç»“æ„, ä¾‹å¦‚å‰è¿°çš„Swiss-rollæ•°æ®é›†;
  - t-SNEæ›´é€‚åˆå­¦ä¹ æ¥è‡ªè‹¥å¹²ç»„å½¼æ­¤ç›¸éš”è¾ƒè¿œçš„æµå½¢çš„æ··åˆæ•°æ®(é€‚åˆæ‰¿æ¥èšç±»ä»»åŠ¡ä½œä¸ºä¸‹æ¸¸), ä¾‹å¦‚æ‰‹å†™æ•°å­—å›¾ç‰‡çš„èšç±».

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

n_samples, random_state = 1500, 170
X_blobs, y_blobs = make_blobs(n_samples=n_samples, random_state=random_state)
plt.scatter(X_blobs[:, 0], X_blobs[:, 1], c=y_blobs)
plt.title("Mixture of Gaussian blobs")
plt.show()
```

## k-meansæ–¹æ³•

$\quad$è€ƒè™‘å°†$N$ä¸ªæ ·æœ¬åˆ†å…¥$K$ä¸ª**ç°‡**(cluster)çš„æŸç§æ–¹æ³•, ä½¿å¾—æ¯ç°‡å†…çš„æ ·æœ¬å°½å¯èƒ½åœ°æ¥è¿‘ç°‡ä¸­å¿ƒ$\boldsymbol{\mu}$. æˆ‘ä»¬äºæ˜¯å¯ä»¥å®šä¹‰ä¸‹è¿°çš„ç•¸å˜å‡½æ•°ä½œä¸ºä¼˜åŒ–ç›®æ ‡:
$$
\mathcal{L}(\mathbf{M}, \mathbf{Z}) := \sum_{n=1}^N{\|\mathbf{x}^{(n)} - \boldsymbol{\mu}_{z_n}\|^2} = \|\mathbf{X} - \mathbf{Z}\mathbf{M}^\mathrm{T}\|_F^2,
$$
å…¶ä¸­,
- çŸ©é˜µ$\mathbf{X} \in \mathbb{R}^{N \times D}$ä¸ºåŸå§‹æ•°æ®;
- ç”±ç°‡ä¸­å¿ƒç»„æˆçš„çŸ©é˜µ$\mathbf{M} \in \mathbb{R}^{D \times K}$, ç¬¬$k$åˆ—å¯¹åº”ç¬¬$k$ä¸ªç°‡çš„ä¸­å¿ƒ$\boldsymbol{\mu}_k$;
- ç¬¬$n$ä¸ªæ ·æœ¬çš„ç±»åˆ«å½’å±è®°ä¸º$z_n$, è‹¥ç‹¬çƒ­åŒ–ä¸º$\mathbf{z} \in \{0, 1\}^K$, åˆ™å¯ä»¥ç»„æˆçŸ©é˜µ$\mathbf{Z} \in \{0, 1\}^{N \times K}$.

$\quad$åœ¨k-meansæ–¹æ³•ä¸­, æˆ‘ä»¬è¿­ä»£åœ°ä¼˜åŒ–ç›®æ ‡å‡½æ•°$\mathcal{L}$. æ¯è½®è¿­ä»£ä¸­ä¾æ¬¡å®Œæˆå¦‚ä¸‹ä¸¤æ­¥:
- ç°‡ä¸­å¿ƒçš„è®¡ç®—
  $$
  \boldsymbol{\mu}_k^* = \arg{\min_{\boldsymbol{\mu}}{\sum_{n: z_n=k}{\|\mathbf{x}^{(n)} - \boldsymbol{\mu}\|^2}}} = \frac{1}{N_k}\sum_{n:z_n=k}{\mathbf{x}^{(n)}};
  $$
- ç°‡çš„åˆ†é…(äº‹å®ä¸ŠåŸºäºè´ªå¿ƒç®—æ³•, greedy algorithm)
  $$
  z_n^* = \arg{\min_k{\|\mathbf{x}^{(n)} - \boldsymbol{\mu}_{z_n}\|^2}}.
  $$

å¯ä»¥è¯æ˜, ä»¥ä¸Šæ¯è½®è¿­ä»£éƒ½ä½¿$\mathcal{L}$å•è°ƒé€’å‡, æœ€ç»ˆæ”¶æ•›åˆ°ä¸€ä¸ª(å±€éƒ¨)æå°ç‚¹. å®è·µä¸­å¯ä»¥å¤šä½œå‡ æ¬¡åˆå§‹åŒ–, å–ç•¸å˜å‡½æ•°æœ€å°çš„é‚£ç»„è§£è¾“å‡º.

$\quad$ä¸èšç±»æœ‰å…³çš„ç®—æ³•å°è£…äº[`sklearn.cluster`](https://scikit-learn.org/stable/modules/classes.html#module-sklearn.cluster)æ¨¡å—ä¸­, è€Œè°ƒç”¨k-meansç®—æ³•çš„æ¥å£ä¸º[`KMeans`](https://scikit-learn.org/stable/modules/generated/sklearn.cluster.KMeans.html#sklearn-cluster-kmeans). æ¨¡å‹åˆ›å»ºéœ€è¦ä¼ å…¥å‚æ•°`n_clusters`(é»˜è®¤ä¸º8), è®­ç»ƒç”¨`fit()`æ–¹æ³•, ç±»åˆ«æ ‡ç­¾åˆ†é…ä¸º`predict()`æ–¹æ³•.
- ç°‡ä¸­å¿ƒç”±å±æ€§`cluster_centers_`è®¿é—®, è¿™ç»™å‡ºå½¢çŠ¶ä¸º`(n_clusters, n_features)`çš„çŸ©é˜µ;
- è®­ç»ƒé›†ä¸Šå„ä¸ªæ ·æœ¬çš„ç±»åˆ«æ ‡ç­¾ç”±å±æ€§`labels_`è®¿é—®.

```python
n_clusters = 3
y_blobs_pred = KMeans(n_clusters=n_clusters, random_state=random_state).fit_predict(X_blobs)
plt.scatter(X_blobs[:, 0], X_blobs[:, 1], c=y_blobs_pred)
plt.title("K-means clustered Gaussian blobs")
plt.show()
```

## è¾¹ç•Œè½¯åŒ–: Gaussianæ··åˆæ¨¡å‹(GMM)

$\quad$k-meansç®—æ³•äº‹å®ä¸Šæ˜¯**æœŸæœ›-æœ€å¤§åŒ–(Expectation-Maximization, EM)ç®—æ³•**çš„ç‰¹ä¾‹. EMæ˜¯ä¸€ç§æ±‚è§£**è€ƒè™‘éšå˜é‡çš„æå¤§ä¼¼ç„¶ä¼°è®¡é—®é¢˜**çš„è¿­ä»£ç®—æ³•, æ¯ä¸ªè¿­ä»£æ­¥$t$ä¾æ¬¡å®Œæˆä¸¤ä¸ªå­æ­¥éª¤:
- **æœŸæœ›æ­¥(E)**, å°†éšå˜é‡$\mathbf{Z}$åœ¨è”åˆåˆ†å¸ƒ$\{\mathbf{X}, \mathbf{Z}\}$çš„å¯¹æ•°ä¼¼ç„¶ä¸­çº¦åŒ–æ‰:
  $$
  Q(\boldsymbol{\theta} | \boldsymbol{\theta}^{(t)}) := \mathbb{E}_{\mathbf{Z} \sim p(\cdot|\mathbf{X}, \boldsymbol{\theta}^{(t)})}[
  \log{p(\mathbf{X}, \mathbf{Z} | \boldsymbol{\theta})}];
  $$
- **æœ€å¤§åŒ–æ­¥**, ä¼˜åŒ–$Q$å€¼:
  $$
  \boldsymbol{\theta}^{(t+1)} := \arg{\max_{\boldsymbol{\theta}}{Q(\boldsymbol{\theta} | \boldsymbol{\theta}^{(t)}).
  }}
  $$

å¯ä»¥è¯æ˜, $\mathbf{X}$çš„å¯¹æ•°ä¼¼ç„¶å‡½æ•°$\log{p(\mathbf{X} | \boldsymbol{\theta}^{(t)})}$éšç€è¿­ä»£è½®æ¬¡$t$çš„å¢åŠ è€Œå•è°ƒé€’å¢. ä¸åŒçš„EMç®—æ³•æ¡ˆä¾‹å¯¹åº”çš„æ˜¯ä¸åŒçš„éšå˜é‡æ¡ä»¶åˆ†å¸ƒ$p(\mathbf{Z} | \mathbf{X},\boldsymbol{\theta})$. åœ¨k-meansé—®é¢˜ä¸­, æ¡ä»¶åˆ†å¸ƒæ˜¯â€œç‹¬çƒ­â€çš„â€œç¡¬â€åˆ†å¸ƒ, å³:
$$
p\left(z_n = k \middle| \mathbf{x}^{(n)}, \{\boldsymbol{\mu}_{k'}\}_{k'=1}^K\right) = \begin{cases}
1, & k=\arg{\min_{k'}{\|\mathbf{x}^{(n)} - \boldsymbol{\mu}_{k'}\|^2}},\\
0, & \mathrm{else}.
\end{cases}
$$

$\quad$è€Œåœ¨**Gaussianæ··åˆæ¨¡å‹**(Gaussian mixture model, GMM)ä¸­, æˆ‘ä»¬é‡‡å–â€œè½¯åˆ†ç±»â€ç­–ç•¥, å°†æ¡ä»¶åˆ†å¸ƒå»ºæ¨¡ä¸º(å¸¦æƒ)Gaussianåˆ†å¸ƒ:
$$
p\left(z_n = k \middle| \mathbf{x}^{(n)}, \{p_{k'}, \boldsymbol{\mu}_{k'}, \mathbf{\Sigma}_{k'}\}_{k'=1}^K\right) \propto p_k\mathcal{N}(\mathbf{x}^{(i)}|\boldsymbol{\mu}_k, \mathbf{\Sigma}_k),
$$
å¹¶ä»¥åŒæ ·çš„è¿­ä»£ç­–ç•¥æ±‚è§£æœ€ä¼˜å‚æ•°$\boldsymbol{\theta}^* \equiv \{p_k^*, \boldsymbol{\mu}_k^*, \mathbf{\Sigma}_k^*\}_{k=1}^K$. ç›¸æ¯”äºk-meansåˆ†é…ç°‡ä¸­å¿ƒçš„åšæ³•, GMMå®è´¨æ˜¯ç»™æ¯ä¸ªæ ·æœ¬ç‚¹åˆ†é…ä¸€ä¸ªGaussianåˆ†å¸ƒçš„å½’å±.

$\quad$è°ƒç”¨GMMç®—æ³•çš„æ¥å£å°è£…äº[`sklearn.mixture.GaussianMixture`](https://scikit-learn.org/stable/modules/generated/sklearn.mixture.GaussianMixture.html#sklearn-mixture-gaussianmixture), è®­ç»ƒä¸é¢„æµ‹çš„æ¥å£ä¾ç„¶ä¸º`fit()`ä¸`predict()`. å‚æ•°`n_components`é»˜è®¤å€¼ä¸º1.
- æœ€ä¼˜å‚æ•°$\{p_k^*\}$ã€æœ€ä¼˜å‡å€¼$\{\boldsymbol{\mu}_k^*\}$ã€æœ€ä¼˜åæ–¹å·®$\{\mathbf{\Sigma}_k^*\}$åˆ†åˆ«å¯ç”±å±æ€§`weights_`ã€`means_`ã€`covariances_`è®¿é—®;
- æˆ‘ä»¬å¯ä»¥å°†k-meansçœ‹ä½œæ˜¯GMMåœ¨å„ä¸ªæƒé‡$p_k$ä¸Gaussianç»„åˆ†æ–¹å·®$\mathbf{\Sigma}_k$å½¼æ­¤ç›¸ç­‰æ—¶çš„ç‰¹ä¾‹. å½“æ•°æ®å‘ˆç°å„å‘å¼‚æ€§æ—¶, k-meansä¸å†é€‚ç”¨, ä½†GMMä»ç„¶å¯ä»¥è€ƒè™‘.

```python
import numpy as np

transformation = [[0.60834549, -0.63667341], [-0.40887718, 0.85253229]]
X_blobs_aniso = np.dot(X_blobs, transformation)
plt.scatter(X_blobs_aniso[:, 0], X_blobs_aniso[:, 1], c=y_blobs)
plt.title("Anisotropic Gaussian blobs")
plt.show()
```

```python
# bad attempt: k-means (due to anisotropic distribution of data)

y_aniso_kmeans = KMeans(n_clusters=n_clusters, random_state=random_state).fit_predict(X_blobs_aniso)
plt.scatter(X_blobs_aniso[:, 0], X_blobs_aniso[:, 1], c=y_aniso_kmeans)
plt.title("K-means clustered anisotropic Gaussian blobs")
plt.show()
```

```python
from sklearn.mixture import GaussianMixture

y_aniso_gmm = GaussianMixture(n_components=n_clusters, random_state=random_state).fit_predict(X_blobs_aniso)
plt.scatter(X_blobs_aniso[:, 0], X_blobs_aniso[:, 1], c=y_aniso_gmm)
plt.title("GMM clustered anisotropic Gaussian blobs")
plt.show()
```
