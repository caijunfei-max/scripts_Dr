```python
result = pd.read_csv("test_results.csv")
result
```

```python
error_list = []
for i in range(79):
    truevalue = result.loc[i, 'True formation energy']
    predictedvalue = result.loc[i, 'Predicted formation energy']
    error = truevalue - predictedvalue
    error_list.append(error)
result['MAE'] = error_list
```

```python
result
```

```python
from figrecipes import PlotlyFig
```

```python
pf = PlotlyFig(result, y_title="predicted formation energy(eV/atom)", x_title="True formation energy(eV/atom)", filename="GNN regression")
pf.xy(('Predicted formation energy', 'True formation energy'), colors='MAE', colorscale="Picnic")
```
